```python
document1 = {
    "Titulo": "Culinaria",
    "Conteúdo" : "Uma receita de bolo de ameixa seca, com chocolate belga e creme de cereja"
}
document2 = {
    "Titulo": "Aquecimento Global",
    "Conteúdo" : "O aquecimento global está afetando muitas areas da economia"
}
document3 = {
    "Titulo": "Cachorrinhos Fofos",
    "Conteúdo" : "Princesa, uma vira-lata, que foi resgatada de umas das encheentes no RS"
}
documents = [document1, document2, document3]
```

```python
data_frame = pd.DataFrame(documents)
data_frame.columns = ["titulo", "conteudo"]
data_frame
```

```python
model =  "models/embedding-001"
```

```python
def embed_fn(title, text):
  return genai.embed_content(model = model, content= text, title = title, task_type= "RETRIEVAL_DOCUMENT")["embedding"]
```

```python
data_frame["embeddings"] = data_frame.apply(lambda row: embed_fn(row ["titulo"], row ["conteudo"]), axis = 1)
data_frame
# apply -> vai criar uma nova coluna no data_frame
# função lambda vai passar por todas as linhas do data_frame
```

```python
def gerar_e_buscar_consulta(consulta, base, model):
  embbeding_consulta = genai.embed_content(model = model, content= consulta, task_type= "RETRIEVAL_QUERY")

  produtos_escalares = np.dot(np.stack(data_frame["embeddings"]), embbeding_consulta["embedding"])
  indice  = np.argmax(produtos_escalares)
  return data_frame.iloc[indice]["conteudo"]
```

```python
consulta = "Qual o nome do cachorro que foi salvo na enchente do Rio Grande do Sul"

resposta = gerar_e_buscar_consulta(consulta, data_frame, model )
resposta
```
